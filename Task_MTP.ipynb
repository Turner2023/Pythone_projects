### Расчёт среднего времени

```python
times_trip = '0 days 00:00:00'
n = 0

for m in machine:
    for r in trip_num:
        shed_fact_cur = shedule_fact.query('trip_num == @r & grafic == @m')
        final_stop = shed_fact_cur['stop_sequence'].max()
        start_stop = shed_fact_cur['stop_sequence'].min()
        if final_stop == 15 and start_stop == 1:
            t2 = shed_fact_cur.loc[shed_fact_cur['stop_sequence'] == final_stop, 'arrival_time'].tolist()
            t1 = shed_fact_cur.loc[shed_fact_cur['stop_sequence'] == start_stop, 'departure_time'].tolist()
            delta = t2[0] - t1[0]
            times_trip = times_trip + delta
            n = n+1

print(times_trip)
print(n)
```

```python
print('Среднее время следования', (times_trip/n))
```

### Определение участка маршрута, на котором наибольшая задержка.

#### Объединение таблиц shedule_plan и shedule_fact.

```python
fact_plan = shedule_plan.merge(shedule_fact, on=['date', 
                                                 'route_id', 
                                                 'grafic', 
                                                 'trip_num', 
                                                 'stop_id', 
                                                 'stop_sequence', 
                                                 'trip_id'], 
                               how='inner')

fact_plan = fact_plan.rename(columns={'arrival_time_x' : 'arrival_plan',
                                      'arrival_time_y' : 'arrival_fact',
                                      'departure_time_x' : 'departure_plan',
                                      'departure_time_y' : 'departure_fact'})

display(fact_plan.head())
```

#### Добавим столбец с временем задержки по отправлению

```python
fact_plan['delay'] = fact_plan['departure_fact'] - fact_plan['departure_plan']
display(fact_plan.head())
```

```python
stages = fact_plan.groupby('stop_sequence', as_index=False)['delay'].mean()
display(stages)
```

```python
display(stages.loc[stages['delay'] == stages['delay'].max(),:])
```

### Метрика пунктуальности для каждого из трёх ТС.
#### Среднее куммулятивное отклонение за один рейс - накопленное абсолютное отклонение от расписания по всем остановкам за рейс и усреднённое по всем рейсам каждого ТС.

```python
# расчёт абсолютного отклонения

def absolute(value):
    t0 = pd.Timedelta(value='0 days 00:00:00')
    if value < t0:
        return (t0 - value)
    else:
        return value

fact_plan['abs_dev'] = fact_plan['delay'].apply(absolute)
display(fact_plan.head())
```

#### Суммарное абсолютное отклонение для всех ТС и рейсов.

```python
abs_group = fact_plan.groupby(by=['grafic', 'trip_num'], as_index=False)['abs_dev'].sum()
display(abs_group.head())
```

#### Абсолютное куммулятивное отклонение за рейс, усреднённое по количеству выходов для каждого маршрута.

```python
abs_group_2 = abs_group.groupby(by='grafic', as_index=False)['abs_dev'].mean()
display(abs_group_2)
```
